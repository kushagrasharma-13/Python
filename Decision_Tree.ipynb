```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
# Load "Iris.csv" dataset
dataset = pd.read_csv('/content/drive/MyDrive/Python/Datasets/Iris.csv')
dataset
```

```python
dataset.info()
```

```python
dataset.describe()
```

```python
corr = dataset.corr()
corr.style.background_gradient(cmap = 'coolwarm')
```

```python
ele = ["Iris-setosa", "Iris-versicolor", "Iris-virginica"]
```

# Length:

Sepal Length -->

```python
plt.figure(figsize=(8,8))
plt.hist(dataset['SepalLengthCm'], color='yellow')
plt.title('Distribution of Sepal Length')
plt.xlabel('Sepal Length')
plt.show()
print("------------------------------------------------------------------------")

x_sl = dataset[['SepalLengthCm']]
y_sl = dataset[['Species']]
x_sl_train, x_sl_test, y_sl_train, y_sl_test = train_test_split(x_sl, y_sl, test_size=0.25, random_state=3)

DTC_sl = DecisionTreeClassifier()
DTC_sl.fit(x_sl_train, y_sl_train)
y_sl_pred = DTC_sl.predict(x_sl_test)

print("Classification Report: ")
print(classification_report(y_sl_test, y_sl_pred))
print("------------------------------------------------------------------------")


print("Accuracy: {:.2f}%".format(accuracy_score(y_sl_test, y_sl_pred)*100))
print("------------------------------------------------------------------------")

print("Confusion Matrix: ")
cm_sl = confusion_matrix(y_sl_test, y_sl_pred)
sns.heatmap(cm_sl,annot=True, xticklabels=ele, yticklabels=ele)
```

Petal Length -->

```python
plt.figure(figsize=(8,8))
plt.hist(dataset['PetalLengthCm'], color='yellow')
plt.title('Distribution of Petal Length')
plt.xlabel('Petal Length')
plt.show()
print("------------------------------------------------------------------------")

x_pl = dataset[['PetalLengthCm']]
y_pl = dataset[['Species']]
x_pl_train, x_pl_test, y_pl_train, y_pl_test = train_test_split(x_pl, y_pl, test_size=0.25, random_state=3)

DTC_pl = DecisionTreeClassifier()
DTC_pl.fit(x_pl_train, y_pl_train)
y_pl_pred = DTC_pl.predict(x_pl_test)

print("Classification Report: ")
print(classification_report(y_pl_test, y_pl_pred))
print("------------------------------------------------------------------------")


print("Accuracy: {:.2f}%".format(accuracy_score(y_pl_test, y_pl_pred)*100))
print("------------------------------------------------------------------------")

print("Confusion Matrix: ")
cm_pl = confusion_matrix(y_pl_test, y_pl_pred)
sns.heatmap(cm_pl,annot=True, xticklabels=ele, yticklabels=ele)
```

Petal and Sepal Length

```python
plt.figure(figsize=(8,8))
plt.subplot(1,2,1)
plt.hist(dataset['SepalLengthCm'], color='yellow')
plt.title('Distribution of Sepal Length')
plt.xlabel('Sepal Length')
plt.subplot(1,2,2)
plt.hist(dataset['PetalLengthCm'], color='red')
plt.title('Distribution of Petal Length')
plt.xlabel('Petal Length')
plt.show()
print("------------------------------------------------------------------------")


x_pl_sl = dataset[['PetalLengthCm', 'SepalLengthCm']]
y_pl_sl = dataset[['Species']]

x_pl_sl_train, x_pl_sl_test, y_pl_sl_train, y_pl_sl_test = train_test_split(x_pl_sl, y_pl_sl, test_size=0.25, random_state=3)

DTC_pl_sl = DecisionTreeClassifier()
DTC_pl_sl.fit(x_pl_sl_train, y_pl_sl_train)
y_pl_sl_pred = DTC_pl_sl.predict(x_pl_sl_test)

print("Classification Report: ")
print(classification_report(y_pl_sl_test, y_pl_sl_pred))
print("------------------------------------------------------------------------")


print("Accuracy: {:.2f}%".format(accuracy_score(y_pl_sl_test, y_pl_sl_pred)*100))
print("------------------------------------------------------------------------")

print("Confusion Matrix: ")
cm_pl_sl = confusion_matrix(y_pl_sl_test, y_pl_sl_pred)
sns.heatmap(cm_pl_sl, annot=True, xticklabels=ele, yticklabels=ele)
```
