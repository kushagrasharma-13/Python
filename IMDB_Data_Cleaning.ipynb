```python
import numpy as np
import pandas as pd
from datetime import date
from statistics import mode
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

# a. Impute the columns containing more than 100 NaN values with suitable central tendency measure.

 ***director_name***, ***plot_keywords*** and can not be imputed due to them being strings

```python
mean_like = dataset['director_facebook_likes'].mean()
dataset['director_facebook_likes'].fillna(mean_like, inplace=True)
```

```python
gross_mean = dataset['gross'].mean()
dataset['gross'].fillna(gross_mean, inplace=True)
```

```python
label_encoder = LabelEncoder()

label_endcoded_rating = label_encoder.fit_transform(dataset['content_rating'])

mode_of_rating = mode(label_endcoded_rating)

mode_rating = mode_of_rating.reshape(-1,1)

s = label_encoder.inverse_transform(mode_rating)
rating = s[0]

dataset['content_rating'].fillna(rating,inplace=True)
```

```python
dataset['budget'].fillna(dataset['budget'].median(), inplace=True)
```

```python
#instead of filling 'title_year' with most frequent year, I will be filling it with current year
current_year = date.today().year
dataset['title_year'].fillna(current_year, inplace=True)
```

```python
mode = dataset['aspect_ratio'].mode()
ratio_mode = mode[0]
dataset['aspect_ratio'].fillna(ratio_mode, inplace=True)
```

```python
dataset.isnull().sum()
```

# b. Remove the rows with NaN values in remaining columns.

```python
dataset = dataset.dropna()
```

```python
dataset.isnull().sum()
```
